```python
#ESCALADO
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = df_combined.drop(columns=['Dx_anemia'])
y = df_combined['Dx_anemia']
```

```python
# Ver qué columnas son datetime
print(X.dtypes)

# Extraer info de la fecha
if 'FechaAtencion' in X.columns:
    X['AñoAtencion'] = X['FechaAtencion'].dt.year
    X['MesAtencion'] = X['FechaAtencion'].dt.month
    X = X.drop(columns=['FechaAtencion'])

# Escalar
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
# Dividimos el dataset escalado (X_scaled) y etiquetas (y) en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
#Desbalance de clases
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
#Gráfico comparativo ANTES Y DESPUES de apllicar SMOTE
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x=y_train, palette='Reds')
plt.title('Antes de SMOTE')
plt.subplot(1, 2, 2)
sns.countplot(x=y_train_resampled, palette='Greens')
plt.title('Después de SMOTE')
plt.tight_layout()
plt.show()
```

```python
# Alinear índices
X_scaled = X_scaled.reset_index(drop=True)
y = y.reset_index(drop=True)

# Tomar muestra
X_sample = X_scaled.sample(n=10000, random_state=42)
y_sample = y.iloc[X_sample.index]

#Configurar validación cruzada
kf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
accuracy_scores = []
log_losses = []

fold = 1
for train_idx, val_idx in kf.split(X_sample, y_sample):
    X_train, X_val = X_sample.iloc[train_idx], X_sample.iloc[val_idx]
    y_train, y_val = y_sample.iloc[train_idx], y_sample.iloc[val_idx]

    # SMOTE para balancear el conjunto de entrenamiento
    smote = SMOTE(random_state=42, k_neighbors=1)
    X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

    #Modelo SVM con kernel lineal
    svm_model = SVC(kernel='linear', probability=True, random_state=42)
    svm_model.fit(X_train_res, y_train_res)

    # Predicción
    y_val_pred = svm_model.predict(X_val)
    y_val_proba = svm_model.predict_proba(X_val)

    acc = accuracy_score(y_val, y_val_pred)
    loss = log_loss(y_val, y_val_proba)

    accuracy_scores.append(acc)
    log_losses.append(loss)
    print(f"Fold {fold} - Accuracy: {acc:.4f}, Log Loss: {loss:.4f}")
    fold += 1
```

```python
plt.figure(figsize=(8, 5))

# Graficar ambas métricas en el mismo eje
plt.plot(range(1, 4), accuracy_scores, marker='o', linestyle='-', color='green', label='Accuracy')
plt.plot(range(1, 4), log_losses, marker='s', linestyle='--', color='red', label='Log Loss')

plt.title('Accuracy y Log Loss por Fold (SVM)')
plt.xlabel('Fold')
plt.ylabel('Valor de la Métrica')
plt.legend()
plt.grid(True)
plt.show()
```
